```python
import matplotlib.pylab as plt
import numpy as np
```

```python
x = np.linspace(-np.pi, np.pi, 201)
print(x)
```

```python
plt.plot(x, np.sin(x))
plt.ylabel("sin(x)")
plt.xlabel("angle [rad]")
plt.axis("tight")
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 4, 1, 5, 2, 6]
plt.plot(x, y, color="blue", linestyle="dashed", linewidth = 3, marker = "o", markerfacecolor="green")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.hist(y)
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```
