```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('loan_data_set.csv')
dataset
```

```python
dataset = dataset.dropna()
dataset
```

```python
gender = {'Male': 1,'Female': 2}
married = {'Yes': 3, 'No': 4}
#dependents = {1: 1, 0: 0, 3+: 3}
education = {'Graduate': 5, 'Not Graduate': 6}
self_employed = {'Yes': 7, 'No': 8}
property_area = {'Semiurban': 9, 'Rural': 10, 'Urban': 11}
loan_status = {'Y': 12, 'N': 13}

dataset.Gender = [gender[item] for item in dataset.Gender]
dataset.Married = [married[item] for item in dataset.Married]
dataset.Education = [education[item] for item in dataset.Education]
dataset.Self_Employed = [self_employed[item] for item in dataset.Self_Employed]
dataset.Property_Area = [property_area[item] for item in dataset.Property_Area]
dataset.Loan_Status = [loan_status[item] for item in dataset.Loan_Status]

print(dataset)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder() 
label_encoder.fit(dataset["Dependents"])

dependents = label_encoder.transform(dataset['Dependents'])
dataset['Dependents'] = dependents
```

```python
dataset.info()
```

```python
dataset
```

```python
dataset = dataset.drop(['Married'], axis=1)
dataset = dataset.drop(['Dependents'], axis=1)
dataset = dataset.drop(['CoapplicantIncome'], axis=1)
dataset
```

```python
X = dataset.iloc[:, 1:9]
Y = dataset.iloc[:, 9]
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
log_classifier = LogisticRegression(random_state=0)
log_classifier.fit(X_train, Y_train)
```

```python
Y_pred = log_classifier.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test, Y_pred)
print(f'Accuracy: {accuracy}')
```
